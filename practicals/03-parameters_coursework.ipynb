In this notebook we will learn about parameters.

````python
# Numerical arrays.
import numpy as np

# Plots.
import matplotlib.pyplot as plt
````

````python
# A range of x values.
x = np.linspace(0.0, 10.0, 21)
x
````

````python
# A range of y values based on x.
y = 4.0 * x - 2.0 + np.random.normal(0.0, 1.0, len(x))
y
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot x vs y.
ax.plot(x, y, 'k.');
````

$ y = a x + b $

````python
# Use x and y to try to find the original parameters.
# params = np.polyfit(x, y, Degree of the fitting polynomial)
# original equation is y = 4.0 * x - 2.0 + np.random.normal(0.0, 1.0, len(x))
params = np.polyfit(x, y, 1)
params

# params is an array- thus below values can be passed into equation of line and plotted
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot x vs y.
ax.plot(x, y, 'k.')

# Plot x vs parameters on x.
# i.e. use the parameters outputted from polyfit to draw the fitting line
# ax.plot(x,y) (y= mx+c)
ax.plot(x, params[0] * x + params[1], 'g-');
````

$ y = a x^2 + b x + c $

````python
# Use x and y to try to find the original parameters.
# y = ax**2 +b*x +c
# params = np.polyfit(x, y, Degree of the fitting polynomial)
params = np.polyfit(x, y, 2)
params
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot x vs y.
ax.plot(x, y, 'k.')

# Plot x vs parameters on x.
ax.plot(x, ((params[0] * x**2) + (params[1] * x) + (params[2])), 'b-');
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot x vs y.
ax.plot(x, y, 'k.')

xtmp = np.linspace(-500.0, 250.0, 1000)

# Plot x vs parameters on x.
ax.plot(xtmp, params[0] * xtmp**2 + params[1] * xtmp + params[2], 'b-');
````

````python
# Let's have a look at plot.
a, b = 1.0, 1.0

# Generate values for x.
x = np.linspace(0.0, 5.0, 101)

# Calculate y.
y = a * np.exp(b * x) + np.random.normal(0.0, 2.0, len(x))

# Create a plot.
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(x, y, 'k.');
````

## Fitting Exponentials

***

https://www.mathcentre.ac.uk/resources/uploaded/mc-bus-loglaws-2009-1.pdf

$ b^c = a \Leftrightarrow \log_b a = c$

$ 2^3 = 8 \Leftrightarrow \log_2 8 =3$

$ y = ae^{bx} $

$ \Rightarrow \log(y) = \log(ae^{bx}) $

$ \Rightarrow \log(y) = \log(a) + \log(e^{bx}) $

$ \Rightarrow \log(y) = \log(a) + bx \log(e) $

$ \Rightarrow \log(y) = \log(a) + bx $

$ \Rightarrow \log(y) = bx + \log(a)$

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, np.log(y), 'k.');
````

````python
# Check which values in log y are nan.
nans = np.isnan(np.log(y))

# Disregard x values where the corresponding log y value is nan.
xnan = x[~nans]

# Disregard y values where the log is nan.
ynan = y[~nans]
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(xnan, np.log(ynan), 'k.');
````

````python
params = np.polyfit(xnan, np.log(ynan), 1)
params
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(xnan, np.log(ynan), 'k.')

ax.plot(xnan, params[0] * xnan + params[1], 'b-');
````

````python
# Undo the effect of taking the log for the parameters.
a_fit, b_fit = np.e**params[1], params[0]
````

````python
# Create a plot.
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, y, 'k.')

ax.plot(x, a_fit * np.exp(b_fit * x), 'b-');
````
